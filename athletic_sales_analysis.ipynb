```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
ath_sales_2020_path = Path('Resources/athletic_sales_2020.csv')
ath_sales_2021_path = Path('Resources/athletic_sales_2021.csv')
ath_sales_2020_df = pd.read_csv(ath_sales_2020_path)
ath_sales_2021_df = pd.read_csv(ath_sales_2021_path)
```

```python
# Display the 2020 sales DataFrame
ath_sales_2020_df
```

```python
# Display the 2021 sales DataFrame
ath_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
data_types_2020 = ath_sales_2020_df.dtypes
print(data_types_2020)
```

```python
# Check the 2021 sales data types.
data_types_2021 = ath_sales_2021_df.dtypes
print(data_types_2021)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

ath_sales_2020_2021_df = pd.concat([ath_sales_2020_df, ath_sales_2021_df],axis="rows", join="inner", ignore_index=True)
ath_sales_2020_2021_df.tail()
```

```python
# Check if any values are null.
check_null_df = ath_sales_2020_2021_df.isnull().values.any()
check_null_df
```

```python
# Check the data type of each column
data_types_2020_2021 = ath_sales_2020_2021_df.dtypes
print(data_types_2020_2021)
```

```python
#Convert the "invoice_date" to a datetime datatype
 
ath_sales_2020_2021_df["invoice_date"] = pd.to_datetime(ath_sales_2020_2021_df["invoice_date"], errors="coerce")
```

```python
#Confirm that the "invoice_date" data type has been changed.
ath_sales_2020_2021_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
df = ath_sales_2020_2021_df
products_sold = df.groupby(['region', 'state', 'city'])['units_sold'].agg(Total_Products_Sold=("sum"))

# Rename the sum to "Total_Products_Sold".
products_sold.sort_values(by=(["Total_Products_Sold"]), ascending=False).head()
# Show the top 5 results.                
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold = ath_sales_2020_2021_df.pivot_table(values='units_sold', index=['region', 'state', 'city'], aggfunc='sum')
products_sold = products_sold.sort_values(by='units_sold', ascending=False)

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold = products_sold.rename(columns={'units_sold':'Total_Products_Sold'})


# Show the top 5 results.
products_sold.head(5)   
```

```python
#Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
#sales_by_women_footwear_date =pd.pivot_table(ath_sales_2020_2021_df, index="invoice_date", values="total_sales", aggfunc='sum')
#sales_by_women_footwear_date = sales_by_women_footwear_date.sort_values(by='product', ascending=False)
sales_by_women_footwear_date = ath_sales_2020_2021_df.pivot_table(values='total_sales', index=['invoice_date'], aggfunc='sum')
#sales_by_women_footwear_date

# Optional: Rename the "total_sales" column to "Total Sales"
sales_by_women_footwear_date.rename(columns={'total_sales':'Total Sales'}, inplace=True)

# Show the table.
sales_by_women_footwear_date.sort_values("Total Sales", ascending= False).head(10)
```

```python
#Resample the pivot table into daily bins, and get the total sales for each day.
daily_sales = sales_by_women_footwear_date.resample('M').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily_sales.resample('M').sum().sort_values("Total Sales", ascending = False).head(10) 
```
